```python
import os
import yaml 

filename = f'{os.getcwd()[:-10]}/config.yaml'
with open(filename) as f:
    dict_yaml = yaml.load(f, Loader=yaml.loader.SafeLoader)
    
dir_source        = dict_yaml['dir_source']
nlead_years       = dict_yaml['number_lead_years']

var               = dict_yaml['variable'][0]
model             = dict_yaml['models_to_train'][0]
verification      = dict_yaml['verification'][0]

y0_clm            = dict_yaml['y0_climatology']
y1_clm            = dict_yaml['y1_climatology']

list_data         = dict_yaml['list_data_options']   

list_2D_measures  = dict_yaml['list_2D_measures']
list_1D_measures  = dict_yaml['list_1D_measures']

domain            = dict_yaml['choose_domain'][0]
resolution        = dict_yaml['choose_common_resolution'][0]

dir_frnt_mdl      = dict_yaml['dir_frnt_mdl']
dir_frnt_obs      = dict_yaml['dir_frnt_obs']
dir_frnt_adj      = dict_yaml['dir_frnt_adj']

dir_frnt_out      = dict_yaml['dir_frnt_diag']
```

```python
import warnings
warnings.filterwarnings('ignore')

import sys
sys.path.insert(1, f'{dir_source}')

from modules_nn.data_load             import (load_verification,
                                              load_forecasts,
                                              load_nan_mask)
from modules_nn.data_preprocessing    import preprocess_data

from modules_aies.info_mdl            import mdl_module    
from modules_aies.info_obs            import obs_module    
from modules_aies.info_var            import var_module    
from modules_aies.info_plt            import plt_module    

from modules_aies.data_load           import (load_nadj_ensemble,
                                              load_xadj_forecasts)      
from modules_aies.data_preprocessing  import align_data_to_common_base  
from modules_aies.data_postprocessing import write_monthly_to_annual    

from modules_aies.util_glbavg         import *                          
from modules_aies.util_unitchg        import UnitChange                 
from modules_aies.util_metrics        import (measure,
                                              get_skill_measures)       

from modules_aies.diagnostics_glbavg  import diagnostics_cascade        
from modules_aies.diagnostics_clim    import diagnostics_climatology_maps 
from modules_aies.diagnostics_skill   import (diagnostics_skill_ts,
                                              diagnostics_skill_maps)    
```

```python
units_PgCyr = r'PgC yr-1'
```

## Set Up ##

```python
dir_mdl_nadj = f'{dir_frnt_adj}/{verification}/{model}/Autoencoder'
dir_mdl_xadj = f'{dir_frnt_adj}/{verification}/{model}' 
```

```python
dir_out            = f'{dir_frnt_out}/ntbk/{resolution}/{domain}/{var}' 
file_out_cascade   = f'ts_cascade_{var}_{verification}'
file_out_clim_ts   = f'ts_clim_{var}_{verification}_base_{y0_clm}-{y1_clm}'
file_out_clim_map  = f'map_clim_{var}_{verification}_base_{y0_clm}-{y1_clm}'
file_out_skill_ts  = f'ts_{var}_{verification}_base_{y0_clm}-{y1_clm}'
file_out_skill_map = f'map_{var}_{verification}_base_{y0_clm}-{y1_clm}'
```

```python
realm = var_module().var_dict[var]['realm']
grid  = var_module().var_dict[var]['grid']
units = var_module().var_dict[var]['units']
```

```python
dict_inf_data = {}
dict_inf_data[model] = mdl_module(dir_frnt_mdl).mdl_dict[model] 
dict_inf_data[verification] = obs_module(f'{dir_frnt_obs}').obs_dict[verification][var] 
```

```python
dict_plt_data = plt_module([verification]+[model],
                                 units=units).plot_data_dict
dict_plt_data_PgCyr = plt_module([verification]+[model],
                                 units=units_PgCyr).plot_data_dict
```

```python
y0_trn = dict_inf_data[verification]['y0']
y1_trn = dict_inf_data[verification]['y1']
y0_trn = max([y0_trn,dict_inf_data[model]['y0']])
y1_trn = min([y1_trn,dict_inf_data[model]['y1']])
y0_trg = 1986 # first target year
y1_trg = 2021 # last  target year
y0_ana = 1991 # first analysis year 
y1_ana = 2020 # last  analysis year
```

```python
verbose = True
if verbose:
    print(f"y0 {model}: {dict_inf_data[model]['y0']}")
    print(f"y1 {model}: {dict_inf_data[model]['y1']}")
    print("======")
    print(f"y0_trn: {y0_trn}")
    print(f"y1_trn: {y1_trn}")
    print("======")
    print(f"y0_trg: {y0_trg}")
    print(f"y1_trg: {y1_trg}")
    print("======")
    print(f"y0_ana: {y0_ana}")
    print(f"y1_ana: {y1_ana}")
    print("======")
    print(f"y0_clm: {y0_clm}")
    print(f"y1_clm: {y1_clm}")
    print("======")
```

## Load Data ##

```python
obs_read = load_verification(dict_inf_data[verification]['dir'],
                             two_dim=True)
```

```python
hnd_read = load_forecasts(dict_inf_data[model]['dir'])
hnd_read = UnitChange(hnd_read).Int2Met(var)
```

```python
hnd_nadj_read = load_nadj_ensemble(dir_mdl_nadj, 
                                   file_key='*nn_adjusted*',
                                   EE=1,
                                   emean=True)
```

```python
idisplay = False
if idisplay:
    display(obs_read)
    display(hnd_read)
    display(hnd_nadj_read)
```

## Preprocess Data ##

```python
obs_in, hnd_in = preprocess_data(obs_read,
                                 hnd_read,
                                 nlead_years=nlead_years)
obs_in = obs_in.rename({'lead_time' : 'time'}).squeeze()
hnd_in = hnd_in.rename({'lead_time' : 'time'}).squeeze()
```

```python
obs_in['time'] = np.arange(nlead_years*12)
hnd_in['time'] = np.arange(nlead_years*12)
```

```python
hnd_nadj_in = hnd_nadj_read.sel(year=slice(y0_trn,
                                           y1_trn))
```

```python
idisplay = False
if idisplay:
    display(obs_in.to_dataset())
    display(hnd_in.to_dataset())
    display(hnd_nadj_in)
```

```python
mask_ocean = load_nan_mask(f'{dir_mdl_nadj}/E1',
                           'nan_mask').squeeze()
```

## Set Data Dictionary ##

```python
dict_data = {}
for key in list_data:
    dict_data[key] = {}
    for freq in ['mon', 'ann']:
        dict_data[key][freq] = {}
        for source in [model]+[verification]:
            dict_data[key][freq][source] = {}  
```

```python
dict_data['hnd']['mon'][model] = hnd_in.to_dataset()
dict_data['hnd_nadj']['mon'][model] = hnd_nadj_in
dict_data['obs']['mon'][verification] = obs_in.to_dataset()
```

```python
idisplay = False
if idisplay:
    display(dict_data)
```

## Load Bias/Trend Adj Forecasts -- if available ##

```python
if [key for key in list_data if key[4:] == 'badj']:
    ds = load_xadj_forecasts(dir_mdl_xadj,
                             f"{var}_{realm}_{model.replace('.','-')}",
                             adj_type='Bias_Adjusted')
    dict_data['hnd_badj']['mon'][model] = ds.sel(year=ds.year.values[:-1]).isel(time=slice(0,nlead_years*12))

if [key for key in list_data if key[4:] == 'tadj']:
    ds = load_xadj_forecasts(dir_mdl_xadj,
                             f"{var}_{realm}_{model.replace('.','-')}",
                             adj_type='Trend_Adjusted')
    dict_data['hnd_tadj']['mon'][model] = ds.sel(year=ds.year.values[:-1]).isel(time=slice(0,nlead_years*12))
```

```python
for key in list_data:
    for source in [model]+[verification]:        
        if bool(dict_data[key]['mon'][source]):
            dict_data[key]['mon'][source]['time'] = np.arange(nlead_years*12)
```

## Dictionary For Cascade Plots ##

```python
nlead_years_show = 3
dict_data_cascade = {}
for source in [model]+[verification]:
    dict_data_cascade[source] = {}
    for key in [key for key in list_data if key[:3] == 'obs' or key[:3] == 'hnd']:
        if bool(dict_data[key]['mon'][source]):
            dict_data_cascade[source][key] = area_weighted_avg(UnitChange(dict_data[key]['mon'][source]).toPgCyr(),
                                                               mask=mask_ocean,
                                                               integral=True).sel(time=slice(0,nlead_years_show*12-1))            
                                                               
```

## Convert Monthly to Annual ##

```python
for key in [key for key in list_data if key != 'obs']: 
    if bool(dict_data[key]['mon'][model]):
        dict_data[key]['ann'][model]  = write_monthly_to_annual(dict_data[key]['mon'][model])
dict_data['obs']['ann'][verification] = write_monthly_to_annual(dict_data['obs']['mon'][verification])    
```

## Align Data to Common Period ##

```python
for key in [key for key in list_data if key[:3] != 'fct']:
    for source in [model]+[verification]:
        if bool(dict_data[key]['ann'][source]):
            dict_data[key]['ann'][source]  = align_data_to_common_base(dict_data[key]['ann'][source].sel(year=slice(y0_trg,y1_trg)),
                                                                       nldyr=nlead_years,
                                                                       time_scale=1)
```

## Downstream Data Dictionary ## 

```python
dict_clim = {}
for key in [key for key in list_data if key[:3] != 'fct']:
    dict_clim[key] = {}
    for freq in dict_data[key]: 
        dict_clim[key][freq] = {}
        for source in [model]+[verification]:
            if bool(dict_data[key][freq][source]):
                dict_clim[key][freq][source] = UnitChange(dict_data[key][freq][source].sel(year=slice(y0_clm,
                                                                                                      y1_clm)).mean(dim='year')).togCm2yr1()
```

```python
dict_anom = {}
dict_full = {}
for key in [key for key in list_data if key[:3] != 'fct']: # do hnd
    dict_anom[key] = {}
    dict_full[key] = {}
    for freq in [freq for freq in dict_data[key].keys() if freq != 'mon']:
        dict_anom[key][freq] = {}
        dict_full[key][freq] = {}
        for source in [model]+[verification]:
            if bool(dict_data[key][freq][source]):
                dict_full[key][freq][source] = dict_data[key][freq][source].sel(year=slice(y0_ana,
                                                                                           y1_ana)) # analysis period
                dict_anom[key][freq][source] = UnitChange(dict_data[key][freq][source]).togCm2yr1() - dict_clim[key][freq][source]
```

## Figures ##

- Figure 2 (panels)

```python
diagnostics_cascade([key for key in list_data if key[:3] != 'fct' and key[:3] != 'obs'],
                    dict_data=dict_data_cascade[model] | dict_data_cascade[verification],
                    units=units_PgCyr,
                    ymin=dict_plt_data['fig2']['ymin'],
                    ymax=dict_plt_data['fig2']['ymax'],
                    titles=dict_plt_data['fig2']['title'],
                    label_ts1=dict_plt_data['fig2']['legend']['ts1'],
                    label_ts3=dict_plt_data['fig2']['legend']['ts3'],
                    xx_step=5,
                    y0=y0_trn,
                    show_samples=True,                     
                    shadesize_left=y0_trg-y0_trn+1,
                    shadesize_right=y1_trg-y0_trg,
                    dir_name=dir_out,
                    file_name=f'{file_out_cascade}_{model.replace(".", "-")}',
                    save=True)
```
